```python
import pandas as pd
df=pd.read_csv(r"C:\Users\HP\ML-Math-DSA\Pandas\Tutorial4\weather_data.csv",parse_dates=['day'])
df.set_index("day",inplace=True)
df



```

```python
new_df=df.fillna(0)#fill nan values with 0
new_df
```

```python
new_df=df.fillna({
    "temperature":0,
    "windspeed":0,
    "event":"no event"})
new_df
```

```python
new_df=df.fillna(method='ffill')#this is forward fill .fill values with previous vales inplace of NAN
new_df
```

```python
new_df=df.fillna(method='bfill',axis="columns")
new_df
```

```python
new_df=df.fillna(method='bfill',limit=1)#when you want to put backward fill a limit
new_df
```

```python
new_df=df.interpolate()#make guess NaN value
new_df
```

```python
new_df=df.interpolate(method="time")#makes a guess Nan values which is near to them
new_df
```

```python
new_df=df.dropna()#if you want to drop na values
new_df
```

```python
new_df=df.dropna(how="all")#prints the row which has atleast some data
new_df
```

```python
new_df=df.dropna(thresh=2)#to keep row the row atleast should have 2 valid values
new_df
```

```python
dt=pd.date_range("01-01-2017","01-11-2017")#to create a missing date rows
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
